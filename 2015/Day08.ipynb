# Part 1

```python
file = open('Inputs/input_day08.txt', 'r')


n_chars, n_code = 0, 0
for line in file:
    n_total = len(line[:-1])
    n_code += n_total
    line = line[1:-2]
    n_discard = 0
    i = 0
    while i <= len(line[:-1]):
        if line[i] == "\\":
            if line[i+1] == "\\":
                n_discard += 1
                i += 1
            elif line[i+1] == "x":
                n_discard += 3
                i += 3
            elif line[i+1] == '"':
                n_discard += 1
                i += 1
        i += 1
    n_chars += n_total - n_discard - 2
    

print(f"FINAL ANSWER: {n_code - n_chars}")
```

# Part 2

```python
file = open('Inputs/input_day08.txt', 'r')


n = 0
for line in file:
    n += sum([c == '"' or c == "\\" for c in line[:-1]]) + 2
    

print(f"FINAL ANSWER: {n}")
```
